## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
df = pd.read_csv('CarPrice_Assignment.csv')
df.head()
```

```python
df.info()
```

## 3. EDA (Exploratory Data Analysis)

```python
## a. Dataset Info & Nulls:
```

```python
df.info()
df.isnull().sum()
```

```python
## b. Feature Engineering (e.g., brand from car name):
```

```python
df['brand'] = df['CarName'].apply(lambda x: x.split(' ')[0].lower())
df['brand'].value_counts()
```

```python
## c. Visual Analysis:
```

```python
sns.histplot(df['price'], kde=True)
sns.boxplot(x='fueltype', y='price', data=df)
sns.scatterplot(x='horsepower', y='price', data=df)
```

```python
## d. Correlation:
```

```python
plt.figure(figsize=(10,6))

numeric_df = df.select_dtypes(include=[np.number])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()

```

## 4. Preprocessing

```python
## a. Drop irrelevant columns:
```

```python
df.drop(['car_ID', 'CarName'], axis=1, inplace=True)
```

```python
## b. One-hot encode categorical columns:
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
 ## c. Train-Test Split:
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('price', axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

##  5. Modeling (Prediction)

```python
## a. Train a Random Forest Regressor:
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
```

```python
## b. Predict and Evaluate:
```

```python
y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score:", r2_score(y_test, y_pred))
```

## 6. Visualize Predictions

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.show()
```
